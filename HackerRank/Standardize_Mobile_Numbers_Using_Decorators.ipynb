```python
def wrapper(f):
    def fun(l):
        # complete the function
        l1 = []
        for i in l:
            num = ""
            if i.startswith("+"):
                for j in range(3):
                    num += i[j]
                num += " "
                for k in range(3, len(i)):
                    num += i[k]
                    if k == 7:
                        num += " "
                l1.append(num)
            elif i.startswith("0"):
                num += "+91 "
                for j in range(1, len(i)):
                    num += i[j]
                    if j == 5:
                        num += " "
                l1.append(num)
            else:
                if len(i) == 10:
                    num += "+91 "
                    for j in range(len(i)):
                        num += i[j]
                        if j == 4:
                            num += " "
                    l1.append(num)
                elif len(i) == 12:
                    num += "+"
                    for j in range(len(i)):
                        num += i[j]
                        if j == 1:
                            num += " "
                        elif j == 6:
                            num += " "
                    l1.append(num)
        f(l1)            
    return fun

@wrapper
def sort_phone(l):
    print(*sorted(l), sep='\n')

if __name__ == '__main__':
    l = [input() for _ in range(int(input()))]
    sort_phone(l) 


```
