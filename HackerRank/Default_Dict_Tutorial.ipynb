```python
from collections import defaultdict
n, m = map(int, input().split())
A, B = defaultdict(list), list()

for i in range(n):
    word = input()
    A[word].append(i+1)
    
for j in range(m):
    B.append(input())
    

for k in B:
    if k in A:
        print(" ".join(map(str,A[k])))
    else:
        print(-1)
    
```
