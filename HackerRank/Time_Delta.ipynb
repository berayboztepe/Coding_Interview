```python
# !/bin/python3

import math
import os
import random
import re
import sys
from datetime import datetime, timedelta
import calendar
from dateutil import tz



# Complete the time_delta function below.
def time_delta(t1, t2):
    
    t1 = t1.split(" ")
    t2 = t2.split(" ")

    tm1 = timedelta(hours=int(t1[5][:3]), minutes=int(t1[5][0] + t1[5][3:])).total_seconds()
    tm2 = timedelta(hours=int(t2[5][:3]), minutes=int(t2[5][0] + t2[5][3:])).total_seconds()

    tz1 = tz.tzoffset('UTC', tm1)
    tz2 = tz.tzoffset('UTC', tm2)
    
    
    
    time1 = datetime(int(t1[3]), list(calendar.month_abbr).index(t1[2]), int(t1[1]), int(t1[4].split(':')[0]), int(t1[4].split(':')[1]), int(t1[4].split(':')[2]), tzinfo=tz1)
    time2 = datetime(int(t2[3]), list(calendar.month_abbr).index(t2[2]), int(t2[1]), int(t2[4].split(':')[0]), int(t2[4].split(':')[1]), int(t2[4].split(':')[2]), tzinfo=tz2)    

    

    return str(abs(time1-time2).total_seconds()).split(".")[0]

if __name__ == '__main__':
    

    t = int(input())

    for t_itr in range(t):
        
        t1 = input()

        t2 = input()

        delta = time_delta(t1, t2)
        print(delta)
            

        

    
```
