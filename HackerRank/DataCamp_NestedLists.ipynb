```python
python_students = []
scores = []
for _ in range(int(raw_input())):
    name = raw_input()
    score = float(raw_input())
    python_students.append([name, score])
    scores.append(score)

maxi = max(scores)
mini = min(scores)


for i in scores:
    if i < maxi  and i != mini:
        maxi = i
        
names = sorted([i[0] for i in python_students if maxi in i])

for i in names:
    print(i)
```
