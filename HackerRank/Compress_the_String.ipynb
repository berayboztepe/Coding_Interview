```python
string, tup, i = input(), [], 0


while True:
    counter = 1
    if len(string) == 1:
        tup.append((counter, int(string[0])))
        break
    for j in range(i+1, len(string)):
        if string[j] == string[i]:
            counter += 1 
        else:
            i = j
            break
    tup.append((counter, int(string[j-1])))
    if j+1 == len(string):
        if string[j] == string[j-1]:
            break
        else:
            tup.append((1, int(string[j])))
            break
            
for i in tup:
    print(i, end=" ")
```
