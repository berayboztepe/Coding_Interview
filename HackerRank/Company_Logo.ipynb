```python
#!/bin/python3

import math
import os
import random
import re
import sys
from collections import defaultdict


if __name__ == '__main__':
    s, sets, cont, c = input(), sorted(set(s)), 0, defaultdict(list)
    for i in s:
        if not i in c[s.count(i)]:
            c[s.count(i)].append(i)
    for i, j in sorted(c.items(), reverse=True):
        if not len(j) > 1:
            print(j[0], i)
            cont += 1
        else:
            for k in sorted(j):
                print(k, i)
                cont += 1
                if cont == 3: break
                    
        if cont == 3: break
```
