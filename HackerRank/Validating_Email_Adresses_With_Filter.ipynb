```python
def fun(s):
    # return True if s is a valid email, else return False

    if "@" and "." in s and s.count("@") == 1 and s.count(".") == 1:
        username, website, extension = s.split("@")[0], s.split("@")[1].split(".")[0], s.split("@")[1].split(".")[1]
        if not username == "" or website == "" or extension == 0:
            c1 = (any([(x.isalpha() == False and x.isnumeric() == False and x != "-" and x != "_") for x in username]))
            c2 = (any([(x.isalpha() == False and x.isnumeric() == False) for x in website]))
            c3 = (any([(x.isalpha() == False) for x in extension]))
            c4 = (len(extension) > 3)

            ct = (any([x for x in[c1, c2, c3, c4]]))

            return list(filter(lambda x: x == False, [ct]))
    else:
        return list(filter(lambda x: x, [False]))
        
def filter_mail(emails):
    return list(filter(fun, emails))

if __name__ == '__main__':
    n = int(input())
    emails = []
    for _ in range(n):
        emails.append(input())

filtered_emails = filter_mail(emails)
filtered_emails.sort()
print(filtered_emails)
```
