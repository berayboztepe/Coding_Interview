```python
size1, size2 = input().split()
size1, size2 = int(size1), int(size2)
inp = '.|.'
text = 'WELCOME'
inp1 = []
sub = (size1 // 2) + 1


if (size1 % 2 != 0) and (size2 == (size1*3)):
    for i in range(size1):
        if i == size1 // 2:
            inp1 = inp1[::-1]
            print(text.center(size2, '-'))
        elif i < size1 // 2:
            inp1.append((2 * i + 1) * inp)
            print((inp1[i]).center(size2, '-'))
        else:
            print((inp1[i - sub]).center(size2, '-'))
      
```
