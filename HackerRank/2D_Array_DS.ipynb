```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the hourglassSum function below.
def hourglassSum(arr):
    max_ = -100
    for i in range(4):
        for j in range(4):
            sums = 0
            for k in range(j, j+3):
                sums += arr[i][k]
                sums += arr[i+2][k]
            sums += arr[i+1][j+1]
            if sums > max_:
                max_ = sums
    return max_

if __name__ == '__main__':

    arr = []
    for _ in range(6):
        arr.append(list(map(int, input().rstrip().split())))

    result = hourglassSum(arr)
    print(result)
```
