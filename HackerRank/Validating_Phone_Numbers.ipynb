```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
import re

n = int(input())

for i in range(n):
    a = input()
    
    if len(a) != 10:
        print("NO")
    elif (re.search(r"[^0-9]", a)): print("NO")
    elif (re.match(r"[7 8 9]", a)): print("YES")
    else: print("NO")
```
