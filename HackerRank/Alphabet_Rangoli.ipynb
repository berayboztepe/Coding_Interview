```python
def print_rangoli(size):
    # your code goes here
    alp = 'abcdefghijklmnopqrstuvwxyz'
    words = ''
    for i in range(size):
        words += alp[i]
    words1 = words[::-1]
    k = 1
    
    for i in range(size*2-1):
        if size == 1:
            print(words)
            break
        if i == (size*2-1) // 2:
            print('-'.join(words1).ljust(size*2-1) + '-' + '-'.join(words[1:]).ljust(size*2-3))
        elif i == 0 or i == size*2-2:
            j = 0
            print(words1[:j*2+1].center((2*(len(words) * 2) - 3), '-'))
        else:
            if i > (size*2-1) // 2:
                j = ((size*2-1) // 2) - k
                k += 1
            else:
                j = i 
            print('-'.join(words1[:j]).rjust(size*2-3, '-') + '-' + '-'.join(words[len(words)-(j+1):]).ljust(size*2-1, '-'))
if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```
